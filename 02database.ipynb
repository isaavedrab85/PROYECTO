#Detalles 01DATABASE

Las colulmnas de mi archivo "01database.csv" de base de datos contienen los ID y puntajes variados de pruebas de admisión PAES, diagnóstico y cursos de matemáticas del primer año.

El detalle de cada columna es el siguiente:

### Columna 1: ID de cada estudiante.
### Columna 2: Sexo
*   M = Masculino
*   F = Femenino

### Columna 3: Tipo de Colegio
*   1 = Particular
*   2 = Subencionado
*   3 = Municipal
*   4 = Bicentenerio

### Columna 4: Puntaje Notas Enseñanza Media $(NEM)$.

Este es el puntaje que se asigna al promedio de notas obtenido en los 4 años de enseñanza media.

### Columna 5: Puntaje Ranking $(P_{RK})$.

Este es el puntaje que se asigna a cada estudiante de acuerdo a la posición relativa de sus notas, en la enseñanza media, en comparación con las notas de tres generaciones de alumnos de su mismo colegio.

### Columna 6: Puntaje Comprensión Lectora $(P_{CL})$.

Es el puntaje obtenido por el estudiante en la PAES de Comprensión Lectora.
### Columna 7: Puntaje Matemáticas 1 $(P_{M_{1}})$.

Es el puntaje obtenido por el estudiante en la PAES de Matemáticas 1.

### Columna 8: Puntaje Matemáticas 2 $(P_{M_{2}})$.

Es el puntaje obtenido por el estudiante en la PAES de Matemáticas 2.

### Columna 9: Puntaje Historia $(P_{H})$.

Es el puntaje obtenido por el estudiante en la PAES de Historia.

### Columna 10: Puntaje Ciencias $(P_{C})$.

Es el puntaje obtenido por el estudiante en la PAES de Ciencias.

### Columna 11: Puntaje Promedio $(P_M)$: $$P_M = \dfrac{P_{M_1} + P_{CL}}{2}.$$

Es el promedio de las pruebas de Matemáticas 1 y Comprensión Lectora.

### Columna 12: Puntaje Ponderado PAES para carrera Ingenieria
$$P_P = \text{NEM} \times 0,1 + P_{Rk}\times 0,3 + P_{CL}\times 0,1+ P_{M_1}\times 0,35 + P_{M_2}\times 0,05 + (P_H\quad o \quad P_C)\times 0,1.$$

Este puntaje depende de cada la ponderación que se da a cada variable y que permite ordenar a los postulantes para su posterior matrícula.


### Columna 13: Puntaje (%) Diagnóstico $(P_{Diag})$.

Es el porcentaje de logro obtenido en la prueba de Diagnóstico de Matemáticas que se realiza a los estudiantes nuevos.


### Columna 14: Puntaje Álgebra y Geometría 2024-1 $(P_{A\& G})$.

Es la nota obtenida en el curso de Álgebra y Geometría.
### Columna 15: Puntaje Álgebra y Geometría 2024-2 (repetición) $(\widetilde{P_{AL}})$.

Es la nota obtenida en el curso de Álgebra y Geometría de repetición. Este curso es rendido por estudiantes que reprobaron en el primer semestre.

### Columna 16: Puntaje Cálculo I 2024-1 $(P_{CI})$.

Es la nota obtenida en el curso de Cálculo I.

### Columna 17: Puntaje Cálculo I 2024-2 (repetición) $(\widetilde{P_{CI}})$.

Es la nota obtenida en el curso de Cálculo I de repetición. Este curso es rendido por estudiantes que reprobaron en el primer semestre.


### Columna 18: Puntaje Álgebra Lineal 2024-2 $(P_{AL})$.

Es la nota obtenida en el curso de Álgebra Lineal. Este curso se puede rendir sólo si se aprueba el curso de Álgebra y Geometría el primer semestre.

### Columna 19: Puntaje Cálculo II 2024-2 $(P_{AL})$.

Es la nota obtenida en el curso de Cálculo II. Este curso se puede rendir sólo si se aprueba el curso de Cálculo I el primer semestre.
